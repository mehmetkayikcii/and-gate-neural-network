```python
  import numpy as np
```

```python

def  sigmoid(net):                        #sigmoid fonksiyonunun tanımlanması
    return 1/(1+np.exp(-net))

def sigmoid_derivative(net):                #sigmoid fonksiyonunun türevi
    return sigmoid(net)*(1-sigmoid(net))

epochs = 1000                             #eğitim sayısı
learning_rate = 0.1

inputs = np.array([[0,0],[0,1],[1,0],[1,1]])  #girdi ve çıktıların tanımlanması
outputs = np.array([0,0,0,1])

weights = np.array([0.01,0.03])            #ağırlıkların belirlenmesi (normalde rastgele belirleniyor)
bias = -0.9

for epoch in range(epochs):

    toplam_hata = 0

    for input, gercek_cikis in zip(inputs,outputs):
      #ileri besleme
      net = np.dot(input, weights) + bias
      ag_cikis = sigmoid(net)

      #hata
      hata = gercek_cikis - ag_cikis
      toplam_hata += hata**2

      #geri yayılım
      delta = learning_rate * hata * sigmoid_derivative(net)
      delta_w = delta * input
      weights += delta_w
      bias += delta

    print(f'{epoch} - Epoch: {toplam_hata}')

#Test etmek için modelinn çıktısını kontrol et
print("Testing trained model on AND inputs:")
for input in inputs:
    final_output = sigmoid(np.dot(input, weights) + bias)
    print(f'Input: {input}, Final Output: {final_output}')

```
